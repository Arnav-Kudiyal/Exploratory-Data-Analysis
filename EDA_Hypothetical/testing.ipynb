```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("spotify_data clean.csv")
df.head()
```

```python
df.info()
```

```python
sns.histplot(data=df["artist_followers"], color = "Green")
plt.xlabel("Artist Followers")
plt.ylabel("Frequency")
plt.title("Spotify Artist Followers")
plt.show()
```

```python

sns.histplot(data=df["track_name"], color = "Green")
plt.xlabel("Artist Names")
plt.ylabel("Frequency")
plt.title("Spotify Artist Names")
plt.show()
```
