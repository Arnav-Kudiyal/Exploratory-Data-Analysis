```python
data['days_Active'] = data['days_Active'].dt.days
```

```python
data['Avg_Order_Value'] = data['Total_Amount'] / data['Units_Sold']
```

```python
data['High_Value_Customer'] = (data['Total_Amount'] > data['Total_Amount'].median()).astype(int)
```
